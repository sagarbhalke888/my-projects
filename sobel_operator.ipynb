```python
from scipy.ndimage.filters import convolve
from scipy.signal import convolve2d
from scipy import misc
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = misc.ascent()
```

```python
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
horizontal_edge_filter = np.array([[1, 2, 1],
                     [0, 0, 0],
                     [-1, -2, -1]])
```

```python
verticle_edge_filter = np.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])
```

```python
horizontal_edge=convolve2d(img, horizontal_edge_filter)
plt.imshow(horizontal_edge,cmap='gray')
```

```python
verticle_edge=convolve2d(img, verticle_edge_filter)
plt.imshow(verticle_edge,cmap='gray')
```
